```python
import torch
import torch.nn as nn
import torch.nn.functional as F

bsz = 24
group = 4
seq_len = 97
beam = 256

score_orig = torch.randn( ( bsz, seq_len, group, beam, beam ) )

selected_probs = []

for i in range( seq_len ):
    score = score_orig[:,i,:,:,:]
    B, N, C, W = score.shape
    flat_score = score.permute(0, 3, 1, 2).reshape(-1, C)
    logits = flat_score # B*W*N, C
    probs = F.softmax(logits, dim=-1)     # この softmax は C についての softmax          
    
    _index_flat = torch.multinomial(probs, num_samples=1, replacement=True)  
        
    #probs_full = F.softmax( logits, dim = -1 )　# この softmax は C についｔの softmax
    selected_prob = torch.gather( probs, -1, _index_flat)
    selected_probs.append( selected_prob.view( B, W, N) )

selected_probs = torch.stack( selected_probs, dim = 0 ) # S, B, N, beam
beam_probs = selected_probs # B, S, beam, N
print( "torch.sum( beam_probs, dim = 2 )[0][0][0]:",torch.sum( beam_probs, dim = 2 )[0][0][0])
#beam_probs = beam_probs / torch.sum( beam_probs, dim = 2 ).unsqueeze( dim = 2 ) # B, S, beam, N
beam_probs = F.softmax( beam_probs, dim = 2 ) # B, S, beam, N     一つ一つの B,S,N について、beam についての総和が1の確率 このsoftmax は W についての softmax
print( "torch.sum( beam_probs, dim = 2 )[0][0][0]:",torch.sum( beam_probs, dim = 2 )[0][0][0])
    
```
